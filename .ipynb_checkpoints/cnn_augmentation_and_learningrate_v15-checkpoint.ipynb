###### Define data augmentation layer for CNN

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
def create_data_augmentation(type_RandomFlip, incr_RandomRotation, incr_RandomZoom):
    '''
    Parameters:
    -----------
    type_RandomFlip: String. "horizontal", "vertical", "horizontal_and_vertical"
    incr_RandomRotation: Float
    incr_RandomZoom: Float
    
    Returns:
    -------
    <keras.engine.sequential.Sequential at 0x2095910e100>
    
    
    
    Notes:
    ------
    Example: see page-221 in Chollet-2E
    
    layers.RandomFlip("horizontal") #Applies horizontal flipping to a random 50% of the images that go through it
    
    https://www.tensorflow.org/api_docs/python/tf/keras/layers/RandomFlip:
    String indicating which flip mode to use. Can be "horizontal", "vertical", or "horizontal_and_vertical". 
                  "horizontal" is a left-right flip and "vertical" is a top-bottom flip. Defaults to "horizontal_and_vertical"
    
    
    layers.RandomRotation(0.1): Rotates the input images by a random value in the range [–10%, +10%] (these are 
                                fractions of a full circle—in degrees, the range would be [–36 degrees, +36 degrees])
    
    RandomZoom(0.2): Zooms in or out of the image by a random factor in the range [-20%, +20%]
    
    Future Refs:
    Find: construction image classification paper, hyperparameter tuning on data_augmentation params to deal with small data
    https://pypi.org/project/Augmentor/
    Paper_2016: Adaptive data augmentation for image classification
    Paper_2019: A survey on Image Data Augmentation for Deep Learning. https://journalofbigdata.springeropen.com/articles/10.1186/s40537-019-0197-0
    
    '''
    #data_augmentation = keras.Sequential(
    #    [
    #        layers.RandomFlip("horizontal"),
    #        layers.RandomRotation(0.1),
    #        layers.RandomZoom(0.2),
    #    ]
    #)
    
    data_augmentation = keras.Sequential(
        [
            layers.RandomFlip(type_RandomFlip),
            layers.RandomRotation(incr_RandomRotation),
            layers.RandomZoom(incr_RandomZoom),
        ]
    )
    
    return data_augmentation
```

### Build CNN

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
def create_model():
    
    data_augmentation = keras.Sequential(
        [
            layers.RandomFlip("horizontal"),
            layers.RandomRotation(0.1),
            layers.RandomZoom(0.2),
        ]
    )
    
    inputs = keras.Input(shape=(256, 256, 3))
    #inputs = keras.Input(shape=(180, 360, 3)) # BH_sim
    #inputs = keras.Input(shape=(299, 299, 3))
    x = data_augmentation(inputs)
    #x = layers.Rescaling(1./255)(x)
    x = layers.Conv2D(filters=32, kernel_size=3, activation="relu")(x)
    x = layers.MaxPooling2D(pool_size=2)(x)
    x = layers.Conv2D(filters=64, kernel_size=3, activation="relu")(x)
    x = layers.MaxPooling2D(pool_size=2)(x)
    x = layers.Conv2D(filters=128, kernel_size=3, activation="relu")(x)
    x = layers.MaxPooling2D(pool_size=2)(x)
    x = layers.Conv2D(filters=256, kernel_size=3, activation="relu")(x)
    x = layers.MaxPooling2D(pool_size=2)(x)
    x = layers.Conv2D(filters=256, kernel_size=3, activation="relu")(x)
    x = layers.Flatten()(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(1, activation="sigmoid")(x)
    model = keras.Model(inputs=inputs, outputs=outputs)
    
    return model
```

```python
def model_compile(learning_rate_in):
    
    #model.compile(loss="binary_crossentropy",
    #              optimizer="adam",
    #              metrics=['accuracy', tf.keras.metrics.Recall(), 
    #                       tf.keras.metrics.Precision(), tf.keras.metrics.AUC()]
    #             )
    
    # https://stackoverflow.com/questions/59737875/keras-change-learning-rate
    # optimizer_in = keras.optimizers.Adam(learning_rate=0.001) # 0.01 # learning_rate=0.001
    optimizer_in = keras.optimizers.Adam(learning_rate=learning_rate_in)
    
    model.compile(loss="binary_crossentropy",
                  optimizer=optimizer_in,
                  metrics=['accuracy', tf.keras.metrics.Recall(), 
                           tf.keras.metrics.Precision(), tf.keras.metrics.AUC()]
                 )
```

### Configure computing environment (i.e., number of gpu, etc.)

```python
from tensorflow.python.client import device_lib
devices = device_lib.list_local_devices()

def sizeof_fmt(num, suffix='B'):
    for unit in ['','Ki','Mi','Gi','Ti','Pi','Ei','Zi']:
        if abs(num) < 1024.0:
            return "%3.1f %s%s" % (num, unit, suffix)
        num /= 1024.0
    return "%.1f%s%s" % (num, 'Yi', suffix)

for d in devices:
    t = d.device_type
    name = d.physical_device_desc
    l = [item.split(':',1) for item in name.split(", ")]
    name_attr = dict([x for x in l if len(x)==2])
    dev = name_attr.get('name', 'Unnamed device')
    print(f" {d.name} || {dev} || {t} || {sizeof_fmt(d.memory_limit)}")
```

### Hyperparameter Loop

```python
from datetime import datetime

import warnings
warnings.filterwarnings("ignore")
```

```python
# vec_learning_rate = [0.001, 0.002, 0.003, 0.01, 0.10]
vec_learning_rate = [0.001, 0.01]
```

```python
param_set = 0

df_hist_aggr = pd.DataFrame(columns=['index', 'loss', 
                                     'accuracy', 'recall', 'precision', 'auc',
                                     'val_loss', 'val_accuracy', 'val_recall', 'val_precision','val_auc',
                                     'Param_Set', 'Learning_Rate', 'Timestamp_Saved'
                                    ]
                           )

for lr in vec_learning_rate:
    param_set = param_set + 1
    print(lr)
    model = []
    model = create_model()
    model_compile(learning_rate_in = lr)
    
    history_curve = []
    history_curve = model.fit(X_train, y_train, epochs=30, batch_size=8, validation_data = (X_test,y_test))
    
    df_hist = pd.DataFrame(history_curve.history)
    df_hist = df_hist.reset_index()
    
    df_hist.columns = ['index', 'loss', 
                       'accuracy', 'recall', 'precision', 'auc',
                       'val_loss', 'val_accuracy', 'val_recall', 'val_precision','val_auc']

    df_hist['Param_Set'] = param_set
    df_hist['Learning_Rate'] = lr
    timestamp_saved = str(int(round(datetime.now().timestamp())))
    #df_hist['Timestamp_Saved'] = str(int(round(datetime.now().timestamp())))
    df_hist['Timestamp_Saved'] = timestamp_saved
    
    df_hist_aggr = pd.concat([df_hist_aggr, df_hist])
    
    # save trained model: 
    # save trained CNN weights
    model.save('./saved_cnn/trained_tf_model_galaxy10decals_15aug2023_' + timestamp_saved + '.h5')
    df_hist.to_csv('./saved_cnn/df_hist_' + timestamp_saved + '.csv', index=False)
    
timestamp_saved_aggr = str(int(round(datetime.now().timestamp())))
df_hist.to_csv('./saved_cnn/df_hist_aggr' + timestamp_saved_aggr + '.csv', index=False)
```

```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from tensorflow import keras

# ML model evaluation
from sklearn.metrics import accuracy_score, precision_score
from sklearn.metrics import recall_score, f1_score, roc_auc_score
from sklearn.metrics import confusion_matrix

import warnings
warnings.filterwarnings("ignore")
```

```python
# load CNN to be evaluated
model = []
#model = keras.models.load_model('./saved_cnn/trained_tf_model_galaxy10decals_15aug2023_1691989984.h5')
model = keras.models.load_model('./saved_cnn/trained_tf_model_galaxy10decals_15aug2023_1691990775.h5')

model_learning_rate = model.optimizer.learning_rate.numpy()

print('Learning Rate:' + str(model_learning_rate))
model.summary()
```

```python
def eval_ml_model(label, y_test_in, y_pred_in):
    """
    Evaluate maching learning (ML) model, binary classification
    """
    print(label)
    print(" ")
    
    score_accuracy = accuracy_score(y_test_in, y_pred_in)
    score_precision = precision_score(y_test_in, y_pred_in)
    score_recall = recall_score(y_test_in, y_pred_in)
    score_f1 = f1_score(y_test_in, y_pred_in)
    score_roc_auc = roc_auc_score(y_test_in, y_pred_in)
    score_confmat = confusion_matrix(y_test_in, y_pred_in)
    
    print('Accuracy: ', score_accuracy)
    print('Precision: ', score_precision)
    print('Recall: ', score_recall)
    print('F1: ', score_f1)
    print('ROC_AUC:', score_roc_auc)
    print(score_confmat)
    viz_confusion_matrix(score_confmat)
    print(" ")
    
def viz_confusion_matrix(confmat):
    fig, ax = plt.subplots(figsize=(2.5, 2.5))
    ax.matshow(confmat, cmap=plt.cm.Blues, alpha=0.3)
    for i in range(confmat.shape[0]):
        for j in range(confmat.shape[1]):
            ax.text(x=j, y=i, s=confmat[i, j], va='center', ha='center')

    plt.xlabel('predicted label')
    plt.ylabel('true label')

    plt.tight_layout()
    # plt.savefig('./figures/confusion_matrix.png', dpi=300)
    plt.show()
```
